```python
# preperation 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/crime-features-la-2020-to-present/Crime_Data_Features.csv")

```

```python

# dictionary mapping
crimes = df[['Crm Cd', 'Crm Cd Desc']].drop_duplicates().set_index('Crm Cd')['Crm Cd Desc'].to_dict()
# desire variables for analysis
variables = ['Crm Cd', 'Vict Age', 'Vict Sex', 'Vict Descent']
# pivot table
pivot = df.pivot_table(index='Crm Cd', columns='Vict Sex', values='Vict Age', aggfunc='mean')
# to bar chart
pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.ylabel('Mean victim age')
plt.xlabel('Crime type')
plt.title('Mean victim age by crime type and victim gender')
# legend crime type
plt.legend(title='Crime type', labels=[crimes[x] for x in pivot.index])
# x-axis labels
plt.xticks(range(len(pivot.index)), pivot.index)
# legend for crime type codes
legend_dict = {x: f"{x}: {crimes[x]}" for x in pivot.index}
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label=legend_dict[x], markerfacecolor=plt.get_cmap('Set1')(x % 9)) for x in pivot.index]
plt.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
#print
plt.show()





```

Based on the plot, it looks like the mean victim age for most crime types and victim genders is between 30 and 50 years old. There are a few exceptions to this pattern, such as crime type 600 (sex offenses), which tends to involve younger victims (mean victim age around 20 years old), and crime type 700 (burglary), which tends to involve older victims (mean victim age around 40 years old).
